```python
from googletrans import Translator
import easyocr
from gtts import gTTS
from IPython.display import Audio
```

```python
reader = easyocr.Reader(['hi'])
translator = Translator()
```

```python
import PIL
from PIL import ImageDraw
im = PIL.Image.open('/content/sample_datsets/25.jpg')
im
```

```python
bounds = reader.readtext('/content/sample_datsets/25.jpg', add_margin=0.55, width_ths=0.7, link_threshold=0.8, decoder='beamsearch',blocklist='=-')
bounds
```

```python
def draw_boxes(image, bounds, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(im, bounds)
```

```python
text_list = reader.readtext('/content/sample_datsets/25.jpg', add_margin=0.55, width_ths=0.7, link_threshold=0.8, decoder='beamsearch',blocklist='=-', detail=0)
text_list
```

```python
text_comb=' '.join(text_list)
text_comb
```

```python
print(translator.detect(text_comb))
```

```python
text_en=translator.translate(text_comb, src='hi')
print(text_en.text)
```

```python
hi_tts=gTTS(text_en.text)
hi_tts.save('trans.mp3')
```

```python
Audio('trans.mp3' , autoplay=True)
```

```python
text_ja=translator.translate(text_comb, src='hi',dest='ja')
print(text_ja.text)
```

```python
hi_tts_ja=gTTS(text_ja.text, lang='ja')
hi_tts_ja.save('trans_ja.mp3')
```

```python
Audio('trans_ja.mp3' , autoplay=True)
```
